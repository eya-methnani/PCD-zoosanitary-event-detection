```python
def detect_language_with_langdetect(line):
    from langdetect import detect_langs
    try:
        langs = detect_langs(line)
        if langs:  # Ensure at least one language was detected
            return langs[0].lang, langs[0].prob
        else:
            return "none", 0.0  # Return this if no languages are detected
    except:
        return "err", 0.0  # Return this in case of an exception
```

```python
df['language'], df['probability'] = zip(*df['Article'].apply(detect_language_with_langdetect))
```

```python
import spacy
from nltk.corpus import stopwords
import stanza
```

```python
# Load spaCy models for English and French
nlp_en = spacy.load("en_core_web_sm")
nlp_fr = spacy.load("fr_core_news_sm")
```

```python
# Load Stanza model for Arabic
nlp_ar = stanza.Pipeline(lang='ar')
```

```python
# Load NLTK stop words for English and French
stop_words_en = set(stopwords.words('english'))
stop_words_fr = set(stopwords.words('french'))
```

```python
def preprocess_text(text, language):
    if language == 'en':
        doc = nlp_en(text)
        tokens = [token.lemma_ for token in doc if token.text.lower() not in stop_words_en and token.is_alpha]
    elif language == 'fr':
        doc = nlp_fr(text)
        tokens = [token.lemma_ for token in doc if token.text.lower() not in stop_words_fr and token.is_alpha]
    elif language == 'ar':
        doc = nlp_ar(text)
        tokens = [word.lemma for sent in doc.sentences for word in sent.words if word.text.isalpha()]
        # Note: Adjust stop words removal for Arabic as needed, depending on your list
    else:
        tokens = text.split()  # Fallback for languages without specific processing
    return " ".join(tokens)
```

```python
# Assuming your DataFrame is loaded as `df`
df['processed_text'] = df.apply(lambda row: preprocess_text(row['Article'], row['language']), axis=1)
```

```python
from transformers import GPTNeoForCausalLM, GPT2Tokenizer
import torch
```

```python
model_name = "EleutherAI/gpt-neo-1.3B"  # You can adjust the model size as needed
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPTNeoForCausalLM.from_pretrained(model_name)
```

```python
prompt = """The following are examples of text categories:
Text: "A new disease outbreak in livestock was reported."
Category: événement à risque zoosanitaire
Text: "Historical analysis of farming practices."
Category: ancien événement
Text: "Today's weather forecast predicts rain."
Category: impertinent
"""
```

Use the model to generate a completion for the prompt, which will include the model's "guess" at the correct category:



> The result will include the entire prompt plus the model-generated category. You'll need to parse this output to extract the category.




```python
inputs = tokenizer(prompt, return_tensors="pt", max_length=1024, truncation=True)
output = model.generate(**inputs, max_length=1024, temperature=0.7)
result = tokenizer.decode(output[0], skip_special_tokens=True)
print(result)
```

```python
# Assuming the category is the last line of the output
category_line = result.split('\n')[-1]  # Get the last line of the output
category = category_line.split(":")[-1].strip()  # Extract the category after the colon and strip whitespace
print("Predicted Category:", category)
```

```python
from transformers import pipeline
# Initialisation de la pipeline Zero-shot Classification
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Définir les catégories cibles
categories = ["événement à risque zoosanitaire", "ancien événement", "impertinent"]

# Appliquer la classification Zero-shot aux articles
results = []
for article in df['Article']:  # Remplacer 'text' par le nom réel de votre colonne d'article
    result = classifier(article, candidate_labels=categories, hypothesis_template="Cet article est {}.")
    results.append(result['labels'][0])  # Ajoute le label le plus probable à la liste des résultats

# Ajouter les labels prédits au DataFrame
df['predicted_label'] = results

# Afficher les premiers résultats
print(df.head())
```
