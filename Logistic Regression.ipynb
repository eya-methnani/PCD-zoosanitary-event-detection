```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import  WordNetLemmatizer

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

# Text cleaning
eng_subset['TEXT'] = eng_subset['TEXT'].str.replace('[^a-zA-Z\s]', '').str.lower().apply(word_tokenize)

# Remove stopwords
stop_words = set(stopwords.words('english'))
eng_subset['TEXT'] = eng_subset['TEXT'].apply(lambda x: [word for word in x if word not in stop_words])

# Lemmatization
lemmatizer = WordNetLemmatizer()
eng_subset['TEXT'] = eng_subset['TEXT'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

# Optional: Remove any remaining non-alphabetic characters
eng_subset['TEXT'] = eng_subset['TEXT'].apply(lambda x: [word for word in x if word.isalpha()])
```

```python
ar_subset = df[df['language_encoded'] == 0]
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import  WordNetLemmatizer

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

# Text cleaning
ar_subset['TEXT'] = ar_subset['TEXT'].str.replace('[^a-zA-Z\s]', '').str.lower().apply(word_tokenize)

# Remove stopwords
stop_words = set(stopwords.words('arabic'))
ar_subset['TEXT'] = ar_subset['TEXT'].apply(lambda x: [word for word in x if word not in stop_words])

# Lemmatization
lemmatizer = WordNetLemmatizer()
ar_subset['TEXT'] = ar_subset['TEXT'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

# Optional: Remove any remaining non-alphabetic characters
ar_subset['TEXT'] = ar_subset['TEXT'].apply(lambda x: [word for word in x if word.isalpha()])
```

```python
ar_subset.head()
```

```python
from sklearn.utils import shuffle

combined_df = pd.concat([eng_subset, ar_subset, fr_subset], ignore_index=True)

# Shuffle the combined DataFrame
combined_df_shuffled = shuffle(combined_df)

# Display the shuffled combined DataFrame
combined_df_shuffled.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score



# Convert text data to TF-IDF features
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # You can adjust max_features as needed
X = tfidf_vectorizer.fit_transform(combined_df_shuffled['TEXT'].apply(lambda x: ' '.join(x)))

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, combined_df_shuffled['STATE_encoded'], test_size=0.3, random_state=42)

# Train a logistic regression model
clf = LogisticRegression()
clf.fit(X_train, y_train)

# Predict labels for the test set
predicted = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predicted)
print("Accuracy:", accuracy)
```

```python
from gensim.models import Word2Vec
from sklearn.base import BaseEstimator, TransformerMixin

class Word2VecVectorizer(BaseEstimator, TransformerMixin):
    def __init__(self, vector_size=100, window=5, min_count=1, workers=4):
        self.vector_size = vector_size
        self.window = window
        self.min_count = min_count
        self.workers = workers
        self.model = None

    def fit(self, X, y=None):
        sentences = [text.split() for text in X]
        self.model = Word2Vec(sentences, vector_size=self.vector_size, window=self.window, min_count=self.min_count, workers=self.workers)
        return self

    def transform(self, X):
        vectors = []
        for text in X:
            words = text.split()
            vector = np.zeros(self.vector_size)
            for word in words:
                if word in self.model.wv:
                    vector += self.model.wv[word]
            vectors.append(vector)
        return np.array(vectors)

# Convert text data to Word2Vec features
word2vec_vectorizer = Word2VecVectorizer(vector_size=100)  # You can adjust vector_size as needed
X = word2vec_vectorizer.fit_transform(combined_df_shuffled['TEXT'].apply(' '.join))

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, combined_df_shuffled['STATE_encoded'], test_size=0.3, random_state=42)

# Train a logistic regression model
clf = LogisticRegression()
clf.fit(X_train, y_train)

# Predict labels for the test set
predicted = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, predicted)
print("Accuracy:", accuracy)
```
